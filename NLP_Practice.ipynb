###Tokenization:

```python
from nltk.tokenize import word_tokenize

text = """
Elon Musk announced in America, Ontario, Los Angeles, Lahore in 2022 that Tesla would expand its factories in Germany and China.
The company invested over $5 billion to increase production of electric cars.
Many customers praised Tesla’s innovation, but some critics argued that the prices were too high.
Meanwhile, competitors like Ford and BMW are also investing heavily in electric vehicles.
elon@musk.com and www.spacex.com or 0307-6219020
"""
doc = word_tokenize(text.lower())

tokens = [word for word in doc]
print(tokens)
```

###Stop Word Removal:

```python
from nltk.corpus import stopwords
import string

stop_words = list(stopwords.words("english"))
# print(stop_words)

clear_tokens = [word for word in doc if word not in stop_words and word not in string.punctuation]
print(clear_tokens)
```

###Stemming:

```python
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()

stemmed = [stemmer.stem(stems) for stems in clear_tokens]

print(stemmed)
```

###Lemmatization:

```python
from nltk.stem import WordNetLemmatizer

lemma = WordNetLemmatizer()

lemmatized = [lemma.lemmatize(lemms, pos="v") for lemms in clear_tokens]

print(lemmatized)
```

###Parts Of Speech:

```python
pos_collection = [nltk.pos_tag(clear_tokens)]

print(pos_collection)
```

###Named Entity Recognition:

```python
from nltk import ne_chunk, pos_tag

pos_tags = pos_tag(clear_tokens)
entities = ne_chunk(pos_tags, binary=False)

for subtree in entities:
    if hasattr(subtree, "label"):
        entity_name = " ".join([token for token, pos in subtree.leaves()])
        entity_type = subtree.label()
        print(entity_name, "->", entity_type)
```

###Bag Of Words(BOW):

```python
from sklearn.feature_extraction.text import CountVectorizer

count_vector = CountVectorizer()

bow_matrix =  count_vector.fit_transform(text)
feature_names = count_vector.get_feature_names_out()

print("Vocabulary:\n", count_vector.get_feature_names_out())
print("\nBoW Matrix:\n", bow_matrix.toarray())
```

###Term Frequency - Inverse Document Frequency(TF-IDF):

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vector = TfidfVectorizer()

tfidf_matrix = tfidf_vector.fit_transform(text)
feature_names = tfidf_vector.get_feature_names_out()

print("Vocabulary:\n", count_vector.get_feature_names_out())
print("\nTF-IDF Matrix:\n", tfidf_matrix.toarray())
```

###Word Embeddings:

```python
from gensim.models import Word2Vec
import nltk
from nltk.tokenize import word_tokenize

sentences  = ["Quick brown fox jumps over the lazy dog.",
              "Fast white rabbit hops above the slow turtle."]

tokenized_sentences = [word_tokenize(s.lower()) for s in sentences]

model = Word2Vec(sentences=tokenized_sentences, vector_size=50, window=5, min_count=1, sg=1)

print("Most similar to 'quick':", model.wv.most_similar("quick")[0])

print("Similarity (fox vs dog):", model.wv.similarity("fox", "dog"))

print("Similarity (rabbit vs turtle):", model.wv.similarity("rabbit", "turtle"))
```

```python
import gensim
from nltk.tokenize import sent_tokenize
import nltk

article = lines

sentences = sent_tokenize(article)

processed_data = [gensim.utils.simple_preprocess(sentence) for sentence in sentences]

model = gensim.models.Word2Vec(window=5, min_count=1, workers=4)
model.build_vocab(processed_data, progress_per=10)
model.train(processed_data, total_examples=model.corpus_count, epochs=10)
model.save('s_w2v')
```

#Machine Learning On Some Big Datasets:

```python
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/SMS.csv")

df['spam'] = df['category'].apply(lambda x: 1 if "spam" in x else 0)
df

# df.to_csv("SMS.csv", index=False)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.svm import LinearSVC

X_train, X_test, y_train, y_test = train_test_split(df.message, df.spam, test_size=0.3, random_state=1)

model = Pipeline([
    ("vectorizer", TfidfVectorizer()),
    ("nb", LinearSVC())
])

model.fit(X_train, y_train)
```

```python
SMS_df = pd.read_csv("test_SMS.csv")
# SMS_df
# SMS_df.to_csv("test_SMS.csv", index=False)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cnv = CountVectorizer()
X_train_vec = cnv.fit_transform(X_train.values)
X_test_vec = cnv.transform(X_test.values)

X_train_vec = X_train_vec.toarray()
X_test_vec = X_test_vec.toarray()
```

```python
import tensorflow as tf
import matplotlib.pyplot as plt

tf_model = tf.keras.Sequential([
    tf.keras.layers.Dense(100, activation='relu'),
    tf.keras.layers.Dense(100, activation='relu'),
    tf.keras.layers.Dense(10, activation='sigmoid')
])

tf_model.compile(
    loss=tf.keras.losses.SparseCategoricalCrossentropy(),
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001),
    metrics=['accuracy']
)

history = tf_model.fit(X_train_vec, y_train, epochs=10)

pd.DataFrame(history.history).plot()
```

```python

import numpy as np

SMS_vec = cnv.transform(SMS_df.message)
SMS_vec = SMS_vec.toarray()

# SMS_vec.dtype
pred = tf_model.predict(SMS_vec)

scores = []
for val in pred:
  scores.append((np.mean(val)*2.2).round())

for cat in scores:
  if cat == 1.0:
    print('spam')
  else:
    print('ham')
```
